```python
import pandas as pd
import matplotlib.pyplot as plt

# Replace 'population_by_age_group.csv' with the actual path to your dataset
dataset_path = 'population_by_age_group.csv'

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(dataset_path)

# Display the first few rows of the DataFrame to understand the structure
print(df.head())

# Assuming the column names in your dataset
country_column = 'Country'
total_column = 'Total'
age_65plus_column = '65+'
age_25_64_column = '25-64 years'
age_15_24_column = '15-24 years'
age_5_14_column = '5-14 years'
age_0_4_column = '0-4 years'

# Sort the DataFrame by the total population in descending order for the bar chart
df_sorted = df.sort_values(by=total_column, ascending=False)

# Plot the bar chart for the top 10 countries
plt.figure(figsize=(12, 8))
plt.bar(df_sorted[country_column][:10], df_sorted[total_column][:10], color='skyblue', label='Total Population')
plt.bar(df_sorted[country_column][:10], df_sorted[age_65plus_column][:10], color='orange', label='65+ Years')
plt.bar(df_sorted[country_column][:10], df_sorted[age_25_64_column][:10], color='green', label='25-64 Years')
plt.bar(df_sorted[country_column][:10], df_sorted[age_15_24_column][:10], color='blue', label='15-24 Years')
plt.bar(df_sorted[country_column][:10], df_sorted[age_5_14_column][:10], color='purple', label='5-14 Years')
plt.bar(df_sorted[country_column][:10], df_sorted[age_0_4_column][:10], color='red', label='0-4 Years')

plt.title('Population Distribution by Age Group in Top 10 Countries (2020)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.legend()
plt.show()
```
